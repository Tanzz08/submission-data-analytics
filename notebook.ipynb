### Cleaning Data

#### Membersihkan tabel `geolocation_df`

```python
geolocation_df.head()
```

```python
geolocation_df = geolocation_df.drop_duplicates()
print("Jumlah duplikasi sekarang : ", geolocation_df.duplicated().sum())
```

```python
geolocation_df[geolocation_df.duplicated(keep=False)].sort_values(by=['geolocation_lat', 'geolocation_lng'])
```

```python
geolocation_df.info()
```

**Insight:**
- drop duplicate yang ada 

#### Membersihkan tabel `order_itmes_df`

```python
datetime_columns = ["shipping_limit_date"]

for column in datetime_columns:
    order_items_df[column] = pd.to_datetime(order_items_df[column])

order_items_df.info()
```

**Insight**
- mengubah tipe data ada kolom `shipping_limit_date` menjadi datetime menggunakan pd.to_datetime()

#### Membersihkan tabel `order_payments_df`


```python
order_payments_df.info()
order_payments_df.describe()
```

```python
order_payments_df[order_payments_df.payment_installments == order_payments_df.payment_installments.min()]
```

```python
order_payments_df.payment_installments.replace(order_payments_df.payment_installments.min(), 1, inplace=True) 
```

```python
order_payments_df[order_payments_df.payment_installments == order_payments_df.payment_installments.min()]
```

```python
order_payments_df[order_payments_df.payment_value == order_payments_df.payment_value.min()]
```

```python
orders_with_zero_payment = order_payments_df.groupby('order_id')['payment_value'].sum().reset_index()
orders_with_zero_payment = orders_with_zero_payment[orders_with_zero_payment['payment_value'] == 0]

print(orders_with_zero_payment)
```

```python
zero_payment_orders = order_payments_df.groupby('order_id')['payment_value'].sum().reset_index()
zero_payment_orders = zero_payment_orders[zero_payment_orders['payment_value'] == 0]

# Cek jumlah data dengan payment_type "not_defined"
zero_payment_orders_with_not_defined = order_payments_df[
    (order_payments_df['order_id'].isin(zero_payment_orders['order_id'])) &
    (order_payments_df['payment_type'] == 'not_defined')
]

print(zero_payment_orders_with_not_defined)
print(f"Total orders with only payment_value = 0 and payment_type 'not_defined': {len(zero_payment_orders_with_not_defined)}")
```

```python
order_payments_df = order_payments_df[~order_payments_df['order_id'].isin(zero_payment_orders_with_not_defined['order_id'])]
```

```python
order_payments_df[order_payments_df.payment_value == order_payments_df.payment_value.min()]
```

```python
order_payments_df.describe()
```

**Insight**
- min payment_installment tidak mungkin bernilai 0 jadi saya ubah value nya
- terdapat anomali juga pada payment_value yang nilai nya 0, dan setelah dilihat dapat dipertimbangkan jika memang ada payment_value bernilai 0 jika payment_type menggunakan voucher, dan jika not defined saya putuskan untuk menghapusnya

#### Membersihkan tabel `order_reviews_df`
